```python
from scipy.stats import spearmanr

def spearman_coefficient(ranking1, ranking2):
    """Calculate Spearman's rank correlation coefficient between two rankings."""
    coef, _ = spearmanr(ranking1, ranking2)
    return coef

def average_spearman(rankings):
    """Calculate average Spearman's rank correlation coefficient across multiple raters."""
    n = len(rankings)
    total_coef = 0
    count = 0
    
    for i in range(n):
        for j in range(i+1, n):
            total_coef += spearman_coefficient(rankings[i], rankings[j])
            count += 1
            
    return total_coef / count
```
